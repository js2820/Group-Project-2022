```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 2 plates are stationary with negative pressure gradient.
H = 0.5
K = -0.1  #pascal/m 
mu = 0.001  #chose the approximate viscosity of water at a temperature of 20 degrees Celsius
y = np.linspace(0,H)  #default 50 spaces.
u_1 = K/(2*mu) * y**2 - (K*H)/(2*mu) * y
# plotting the graph.
plt.plot(u_1,y)
plt.xlabel('$u_1$')
plt.ylabel('$x_2$')
plt.show()
```

```python
n1 = 0.8  #n1 < 1 

u1 = (((abs(K))*H)/(2*mu))**(1/n1) * (n1*H)/(2*(1+n1)) * (1 - abs(1-(2*y)/H)**(1+1/n1))

plt.plot(u1,y)

plt.xlabel('$u_1$')
plt.ylabel('$x_2$')
plt.show()
```

```python
n2 = 1
u2 = (((abs(K))*H)/(2*mu))**(1/n2) * (n2*H)/(2*(1+n2)) * (1 - abs(1-(2*y)/H)**(1+1/n2))
plt.plot(u2,y)
plt.xlabel('$u_1$')
plt.ylabel('$x_2$')
plt.show()
```
